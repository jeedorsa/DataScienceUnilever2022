Si se hace una figura y hace esa figura se le agrega un plot 
pero puedo ir agregandole subplot y con eso puedo tener muchos gráficos dentro de una sola figura.
1. Crear la figura 
2. ingresar el codigo .add_subplot() # 1 fila, 1 columna y un gráfico, donde las columnas o filas deben ser iguales para la cantidad de gráficos


```python
### figura1
fig1 = plt.figure(dpi=100)
ax1 = fig1.add_subplot(1,2,1) #1 fila, 1 columna y gráfico 1.
ax1.plot([1,2,3,4], [1,4,9,16], 'o--', color="orchid", markerfacecolor="lightgreen", lw=2, ms=10)
ax1.axis([0,5,0,17])

ax2 = fig1.add_subplot(1,2,2) #1 fila, 1 columna y gráfico 1.
ax2.plot([1,2,3,4], [1,4,9,16], 'o--', color="red", markerfacecolor="yellow", lw=1, ms=10)
ax2.axis([0,5,0,17])
```

# Agregar texto a los gráficos
fig.subtitle -> titulo superior
ax.text -> se le debe dar las coordenadas para ubicar un texto en cualquier coordenada
agregar un marcador-> plt.plot(x,y,'tipo de marcador')
agregar lineas-> .annotate('texto', coordenada a apuntar, coordenadad del texto, arrowprops=dict(facecolor='', shrink=))


```python
fig = plt.figure(dpi=100)
fig.suptitle("Titulo superior en negrita (bold)", fontsize=14, fontweight="bold")

ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85) #ajusta posición de gráfico

ax.set_title("Titulo de ejes, x vs y")
ax.set_xlabel("Etiquta de eje x")
ax.set_ylabel("Etiquta de eje y")

#Texto a todo colooooooor!
ax.text(1.5,8, 'texto a todo color en coordenadas (1.5, 8)', color="darkgreen", fontsize=12)

#Graficamos un pentagono y una linea segmentada
ax.plot([2],[6], 'p', ms=10)
ax.plot([1,2,3,4,5,6,7,8], [0.6,3,2,1,4,5,6,1], 'rd--', lw=2)

# Generamos anotación con una flecha! flecha coordenada (2,1), texto de flecha coordenada (3,1)
ax.annotate('Soy un pentagono indefenso', xy=(2,6), xytext=(4,7), arrowprops=dict(facecolor='black', shrink=0.02))

ax.axis([0,10,0,10])
plt.show()
```

```python
azucar_df=df_vinos.groupby(df_vinos['residual sugar']).nunique()
```

```python
base_ph=df_vinos.groupby(df_vinos['pH']).nunique()
```

```python
fig = plt.figure(dpi=200)
fig.suptitle("Comparación de caracteristicas de vinos", fontsize=14, fontweight="bold")
ax = fig.add_subplot(1,2,1)
ax.set_title("Distribución según pH")
ax.set_xlabel("pH")
ax.set_ylabel("Cantidad")
ax.hist(x = base_ph.index)

ax2 = fig.add_subplot(1,2,2)
ax2.set_title("Distribución según azucar residual")
ax2.set_xlabel("azucar residual")
ax2.hist(x = azucar_df.index,color='cyan')




plt.show()
```

```python
ph_vs_acidez=df_vinos.groupby(df_vinos['pH'])['alcohol'].mean()
ph_vs_acidez
```

```python
valor_medio_sup=df_vinos.loc[df_vinos['pH'].sort_values()>3.31,:]
valor_medio_sup
```
