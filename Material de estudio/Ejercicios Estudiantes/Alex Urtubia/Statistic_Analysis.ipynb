En esta tercera parte del proyecto, el objetivo es continuar indagando en la naturaleza de las tiendas y productos esta vez a nivel estadístico, graficando la información mediante histogramas, diagramas de cajas, mapas de calor, etc para comprobar sus distribuiciones, correlaciones y demás elementos estadísticos que resuman el comportamiento de los datos.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Proyecto1/df_sin_v0.csv')
```

```python
df.head()
```

```python
fig1 = plt.figure(figsize=(20, 8))
fig1.suptitle("Distribución de los pesos de los productos",fontsize=30, fontweight='bold', y=1.05)
plt.style.use('seaborn')


# Se añade un subplot de una fila dos columnas, este subplot es el primero de la fila
ax1 = fig1.add_subplot(1,2,1) #1ra fila, 1ra columna, gráfico 1.
df['Item_Weight'].hist(bins = 20, color='y', edgecolor='g', alpha=.7)
plt.xlabel("Pesos", size=15, labelpad=15)
plt.ylabel("Cantidades", size=15, labelpad=15)
plt.ticklabel_format(useOffset=False, style='plain')
# plt.title("Distribuición general en pesos de productos", size=20, pad=14)

ax2 = fig1.add_subplot(1,2,2) #1ra fila, 1ra columna, gráfico 1.
boxplots = ax2.boxplot(df['Item_Weight'],
           notch = True,
           labels=["Pesos"],
           widths = .7,
           patch_artist=True,
           medianprops = dict(linestyle='-', linewidth=3, color='green'),
           boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'yellow', alpha = .6)
          );
# Cuanto más comprendas cualquier librería, más podrás hacer
boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('yellow')
# plt.title('Pesos de cada producto', fontsize = 20, pad=15);
plt.xticks(fontsize = 16);
plt.yticks(fontsize = 16);
```

```python
it_v100 = df['Item_Visibility']*100

fig1 = plt.figure(figsize=(20, 8))
fig1.suptitle("Porcentaje de visibilidad de los productos",fontsize=30, fontweight='bold', y=1.05)
plt.style.use('seaborn')


# Se añade un subplot de una fila dos columnas, este subplot es el primero de la fila
ax1 = fig1.add_subplot(1,2,1) #1ra fila, 1ra columna, gráfico 1.
it_v100.hist(bins = 20, color='g', edgecolor='k', alpha=.7)
plt.xlabel("% Visibilidad", size=15, labelpad=15)
plt.ylabel("Cantidades", size=15, labelpad=15)
plt.ticklabel_format(useOffset=False, style='plain')
# plt.title("Distribuición en visibilidad de productos", size=20, pad=14)

ax2 = fig1.add_subplot(1,2,2) #1ra fila, 1ra columna, gráfico 1.
boxplots = ax2.boxplot(it_v100,
           notch = True,
           labels=["% Visibilidad"],
           widths = .7,
           patch_artist=True,
           medianprops = dict(linestyle='-', linewidth=3, color='black'),
           boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'yellow', alpha = .6)
          );
# Cuanto más comprendas cualquier librería, más podrás hacer
boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('green')
# plt.title('Pesos de cada producto', fontsize = 20, pad=15);
plt.xticks(fontsize = 16);
plt.yticks(fontsize = 16);

```
