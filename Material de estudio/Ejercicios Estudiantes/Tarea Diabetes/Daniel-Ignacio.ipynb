```python
preg_filter = df["Pregnancies"] < 14
gluc_filter = df["Glucose"] > 0
BP_filter = df["BloodPressure"] > 0
BMI_filter = df["BMI"] > 0
sk_th_filter = df["SkinThickness"] > 0
df1 = df.loc[preg_filter & gluc_filter & BP_filter & BMI_filter & sk_th_filter, :]
df1
```

```python
df1.columns
```



-Para la columna 'Pregnancies' decidimos tomar como valor
máximo 13 embarazos. Ya que estos valores tan altos alejados del promedio pueden corresponder a un error en la toma de muestras.

-Para el caso de la columna 'Glucose', 'BloodPressure' y 'SkinThickness' dejamos como valor mínimo 1 y sacamos las muestras que tenían valores 0, ya que seguramente corresponden a errores en la toma de muestras.
Tenemos dudas respecto a como se realiza la medición para el dataset para estás variables, sobretodo para "BloodPressure" donde hemos visto que la medición se expresa en presión diastolica y sistolica.

-También tenemos dudas respecto a 'Insulin', ya que en la Diabetes tipo 1 las personas producen poco o nada de insulina, pero no sabemos si es que puede llegar a ser 0.
