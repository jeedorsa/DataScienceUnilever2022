```python
import pandas as pd

filename = "/content/diabetes.csv"
df = pd.read_csv(filename)

df.info()
```

```python
# La variable de 'Glucose' y 'BloodPressure' se estimo en valores por sobre cero, debido a que se
# asume que los valores normales considerados para una persona viva no pueden ser menores.
filter1 = df["Glucose"] > 0
filter3 = df["BloodPressure"] > 0
filter4 = df["BMI"] != 0

df1 = df.loc[filter1 & filter3 & filter4, :]
df1

```

```python
# En cuanto al filtro 'Pregnancies', se decidio no filtrar por dicha variable, debido a que en consideracion a otros casos encontrados en el df (personas con mas de 10 embarazos
# y entre los mismos rangos de edad) no se considero 'imposible' que hubieran casos con mas de 14 embarazos. Ademas, considerando la edad promedio de la menarquia (12 anios)
# aquellos casos son perfectamente posibles en relacion a la edad de dichas personas.

print("Aqui podemos ver los casos extremos encontrados en el df:")

p_filter = df['Pregnancies'] >= 14

df1 = df.loc[p_filter, :]
df1
```

```python
print("Aqui podemos ver las edades de los casos con mas de 10 embarazos dentro del data frame")
# Como se puede ver, la cantidad de casos con mas de 10 embarazos no es menor...
p_filter1 = df['Pregnancies'] > 10

df2 = df.loc[p_filter1, 'Age' :]
df2
```

```python
df1.to_csv(path_or_buf='diabetes_filtrado.csv',index = False)
```
