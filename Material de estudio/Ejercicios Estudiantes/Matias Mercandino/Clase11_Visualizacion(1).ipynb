```python

import numpy as np

import matplotlib.pyplot as plot

 

# Get x values of the sine wave

time        = np.arange(0, 20, 0.1);
 

# Amplitude of the sine wave is sine of a variable like time

amplitude   = np.sin(time)

 

# Plot a sine wave using time and amplitude obtained for the sine wave

plot.plot(time, amplitude)

 

# Give a title for the sine wave plot

plot.title('Sine wave')

 

# Give x axis label for the sine wave plot

plot.xlabel('Time')

 

# Give y axis label for the sine wave plot

plot.ylabel('Amplitude = sin(time)')

 

plot.grid(True, which='both')

 

plot.axhline(y=0, color='k')

 

plot.show()

 

# Display the sine wave

plot.show()
```

```python
filename = "/content/19062020_NEM_PERCENTILES_JOVENES_2019_28032020_PUBL.csv"
data3= pd.read_csv(filename,sep=";")
data3.head()
data3.info()
```

```python
orden_nem=data3.sort_values("NEM",ascending=True)
orden_nem.head(10)
```

```python
#orden_nem["NEM"]=="," pasarlo a cadena y con un for recorro todo el pandas y donde haya , ponerle un replace por .
orden_nem["NEM"]=orden_nem['NEM'].astype(str).replace(",", ".").astype(float)
```

```python
orden_nem.groupby(["COD_DEPE"])["NEM"].count()
```

```python
a=orden_nem.groupby(["COD_DEPE"])["NEM"].mean()
plt.figure(dpi=100)
plt.xlabel("CODIGO ESTABLECIMIENTOS")
plt.ylabel("NEM")
plt.title("PROMEDIO NEM SEGUN TIPO DE ESTABLECIMIENTOS")
plt.plot(a)
```
