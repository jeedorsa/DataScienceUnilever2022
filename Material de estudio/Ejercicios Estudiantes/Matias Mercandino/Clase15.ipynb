```python
filename = "/content/heart_2020_cleaned (1).csv"
df = pd.read_csv(filename)
df.head()
```

No existen datos raros

```python
df.describe()
```

No existen datos nulos

```python
df.isna().sum()
```

Creo columna de rangos segun BMI

```python
x2 = []
for i in df["BMI"]:
    if i < 18.5:
        x2.append("insuficiente")
    elif i>18.5 and i <=24.9:
        x2.append("saludable")
    elif i >25.0 and i <=29.9:
        x2.append("sobrepeso")
    else:
        x2.append("obesidad")
```

```python
df["Rangos"]=x2
df.head()
```

```python
#nunqiue que tantos valores diferentes hay 
rango=df.groupby("Rangos")["BMI"].count()
rango
```

```python
import matplotlib.pyplot as plt

#Gráfico de torta; cortes ordenados y gráficados contra reloj.
size = rango.values
etiquetas = rango.index

fig1 = plt.figure(dpi=100)
ax1 = fig1.add_subplot(1,1,1) 

ax1.pie(size, labels=etiquetas, autopct="%.2f%%", shadow = False, startangle=-90)
# autopct = porcentaje automático; shadow = grafico con sombra; startangle = ángulo de inicio
# %.2f%% formato de decimal, viene del lenguaje C. Sí, python está escrito en C.

ax1.axis("equal") #asegura proporcion igualitaria, vale decir, suma de porcentajes 100% (círculo)

plt.show()
```

Grafico BMI promedio segun rango de edad

```python
edad=df.groupby("AgeCategory")["BMI"].mean()
edad
```

```python
sns.barplot(edad.index, edad.values);
plt.grid();
plt.xlabel("Año",fontsize = 25)
plt.ylabel("Peso Promedio",fontsize = 25)
plt.xticks(rotation = 90);
plt.xticks(fontsize = 15);
plt.yticks(fontsize = 15);
```

# Ejercicio 3 

Cargar dataset, hacer limpieza, graficar los 20 estados que mas tienen ventas, crear una nueva columna basado en un profit positivo es ganancia, profit negativo es perdida, graficar con un scatter el comportamiento del profit y ventas en New York. Hacer una curva del comportamiento de los vendedores, ascendentemente.


```python
filename = "/content/Sample - Superstore.csv"
df = pd.read_csv(filename,encoding='ISO-8859-1')
df.head()
```

No existen datos nulos

```python
df.isna().sum()
```

Cargar dataset, hacer limpieza, graficar los 20 estados que mas tienen ventas.

```python
top20estados = df.groupby(['State'])['Sales'].sum().sort_values(ascending = False).head(20)
top20estados
```

```python
sns.barplot(top20estados.index, top20estados.values);
plt.grid();
plt.xlabel("Estados",fontsize = 25)
plt.ylabel("Ventas",fontsize = 25)
plt.title("Top 20 estados segun ventas",fontsize = 25)
plt.xticks(rotation = 90);
plt.xticks(fontsize = 15);
plt.yticks(fontsize = 15);
```

crear una nueva columna basado en un profit positivo es ganancia, profit negativo es perdida, graficar con un scatter el comportamiento del profit y ventas en New York

```python
x3 = []
for i in df["Profit"]:
    if i >=0:
        x3.append("Ganancia")
    else: 
        x3.append("Perdida")
```

```python
df["Ganancia/Perdida"]=x3
df.head()
```

```python
df["City"].value_counts()
```

```python
new_york=(df["City"]=="New York City")
new_york=df[new_york]
new_york.head()
```

Filtrando por Ganancias y Perdidas

```python
ganancias=(new_york["Ganancia/Perdida"]=="Ganancia")
ganancias=new_york[ganancias]
ganancias.head()

perdidas=(new_york["Ganancia/Perdida"]=="Perdida")
perdidas=new_york[perdidas]
perdidas.head()
```

```python
# ¡Juega con este código y mira qué pasa!
# Ajustar el tamaño de la figura
plt.figure(figsize=(10,6))
# Subplot 1 
plt.subplot(1, 2, 1);
plt.scatter(ganancias['Sales'].values, ganancias['Profit'].values, c= 'g');
plt.title('Ganancias en New York', fontsize = 12);
# Subplot 2
plt.subplot(1, 2, 2);
plt.scatter(perdidas['Sales'].values, perdidas['Profit'].values, c= 'r');
plt.title('Perdidas en New York', fontsize = 12);
```

Hacer una curva del comportamiento de los vendedores, ascendentemente.

```python
top20vendedores = df.groupby(['Customer Name'])['Sales'].sum().sort_values(ascending = True).head(20)
top20vendedores
```

```python
plt.plot(top20vendedores.index,top20vendedores.values, label = "Ventas",c = 'magenta', marker = '.', markersize = 20);
plt.legend(loc=(1.02,0))
plt.xlabel("Ventas",fontsize=15);
plt.ylabel("Vendedores",fontsize=15);
plt.title("Ventas por Vendedor",fontsize=25);  
plt.xticks(rotation = 90);
plt.grid()
```
