```python
def letrear(string):
    mayus = "ABCDEFGHIJKLMNOPQRSTUVWXYZÑÁÉÍÓÚ"
    minus = mayus.lower()
    separadas = {'mayúsculas': list(), 'minúsculas': list()}
    for c in string:
        if c in mayus:
            separadas['mayúsculas'].append(c)
        elif c in minus:
            separadas['minúsculas'].append(c)
    
    ya_vistas = list()
    repetidas = int()
    for c in separadas['mayúsculas']:
        if separadas['mayúsculas'].count(c) > 1 and c not in ya_vistas:
            repetidas += 1
            ya_vistas.append(c)
    for c in separadas['minúsculas']:
        if separadas['minúsculas'].count(c) > 1 and c not in ya_vistas:
            repetidas += 1
            ya_vistas.append(c)
    print(f"Hay {repetidas} letras repetidas en la cadena.")
    
    return separadas
```

```python
print(letrear("HOLAAAAAAAAAAAAAAAA como estás? muy bien y tú? ay qué bien me alegro"))
```
