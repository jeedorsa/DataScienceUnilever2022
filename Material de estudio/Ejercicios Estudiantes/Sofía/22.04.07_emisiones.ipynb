# Emisiones contaminantes

# 1. DB setup

Generar un DataFrame con los datos de los cuatro ficheros.

```python
import pandas as pd
```

```python
df16 = pd.read_csv('emisiones-2016.csv', sep=';')
df17 = pd.read_csv('emisiones-2017.csv', sep=';')
df18 = pd.read_csv('emisiones-2018.csv', sep=';')
df19 = pd.read_csv('emisiones-2019.csv', sep=';')
```

```python
df = pd.concat([df16, df17, df18, df19], axis=0)
df
```

Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, etc.

```python
df.columns
```

```python
non_day_cols = ['ESTACION', 'MAGNITUD', 'ANO', 'MES']
day_cols = ['D01', 'D02', 'D03', 'D04',
       'D05', 'D06', 'D07', 'D08', 'D09', 'D10', 'D11', 'D12', 'D13', 'D14',
       'D15', 'D16', 'D17', 'D18', 'D19', 'D20', 'D21', 'D22', 'D23', 'D24',
       'D25', 'D26', 'D27', 'D28', 'D29', 'D30', 'D31']

df = df[non_day_cols + day_cols]
df.head()
```

# 2. Separación columna

Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.

```python
df = df.melt(id_vars=non_day_cols, var_name='DIA', value_name='EMISIONES')
df
```

# 3. Concatenación fecha

Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo datetime).

```python
from datetime import datetime
```

```python
def concatefecha(fila):
    return datetime(f"{fila['ANO']}-{fila['MES']}-{fila['DIA'][1:]}")
df['FECHA'] = df.apply(concatefecha, axis=0)
df
```
