```python
# Pedir número hasta que cumpla condiciones
wrong = True
while wrong:
    num = input('Ingrese un número entero entre 2 y 1000:')
    try:
        num = int(num)
    except TypeError:
        print("No ingresó un número entero.")
        continue
    if not 1 < num < 1001:
        print("Su número no está dentro del rango aceptado")
        continue
    else:
        wrong = False

# Mostrar si es par o impar
answer = ("par" if (num % 2 == 0) else "impar")
print(f"Número {answer}")

# Mostrar si es primo
divisores = list()
for i in range(num, 0, -1):
    if num % i == 0:
        divisores.append(i)
answer = ("Sí" if (divisores == [num, 1]) else "No")
print(f"{answer} es un número primo")

# Mostrar divisores
print("Conjunto de divisores:")
for i in sorted(divisores):
    print(i)
```
