```python
class LibretaDeNotas:
    
    def __init__(self):
        self.alumnos = dict()
        n_alumnos = int(input())
        for i in range(n_alumnos):
            nombre = input(f"Nombre de alumno {i+1}: ")
            self.alumnos[nombre] = list()
            nota = float(input("Ingrese la siguiente nota: "))
            while nota >= 0:
                self.alumnos[nombre].append(nota)
                nota = float(input("Ingrese la siguiente nota: "))
    
    def mostrar_alumnos(self):
        for nombre, notas in self.alumnos.items():
            print(f"Promedio {nombre}: {sum(notas)/len(notas)}")
    
    def invertir_notas(self):
        for nombre, notas in self.alumnos.items():
            self.alumnos[nombre] = reversed(notas)
            print(f"Notas {nombre}: {list(x for x in self.alumnos[nombre])}")
```

```python
libreta = LibretaDeNotas()
```

```python
libreta.mostrar_alumnos()
```

```python
libreta.invertir_notas()
```
