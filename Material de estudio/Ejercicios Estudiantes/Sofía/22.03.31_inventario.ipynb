```python
def agregar_producto(inventario, producto):
    inventario[len(inventario)] = {
        'nombre': producto,
        'descripcion': '',
        'stock': int(),
        'precio': int(),
        'variaciones': ['', '', '']
    }
```

```python
inventario = dict()
producto = input("¿Qué producto desea agregar?")
while producto != "SALIR":
    agregar_producto(inventario, producto)
    producto = input("¿Qué producto desea agregar?")
```
